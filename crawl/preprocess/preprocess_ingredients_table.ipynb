```python
dir_fieldnames = {
    'product_ingredients': ['product_id', 'total_ingredients', 'product_ingredients_description', 'ewg', 'natural'],
    'ingredients_table': ['product_id', 'title', 'introtext', 'cir_rating', 'categories', 'ewg_ingre', 'properties'],
    'harmful': ['product_id', 'type', 'title', 'description', 'list', 'slug'],
    'positive': ['product_id', 'type', 'title', 'description', 'list'],
    'notable': ['product_id', 'type', 'title', 'list'],
}

ingredients_table = pd.read_csv(r'D:\Study\DATN\KG_Cosmetics_Jasmine\data\ingredients\ingredients_table.csv', names=dir_fieldnames['ingredients_table'])
ingredients_table = ingredients_table.drop_duplicates()
```

```python
def preprocess_introtext(text):
    if isinstance(text, str):
        cleaned_text = BeautifulSoup(text, "html.parser").get_text()
        return cleaned_text
    return text
```

```python
ingredients_table.loc[:, 'preprocessed_introtext'] = ingredients_table.loc[:, 'introtext'].apply(preprocess_introtext)
ingredients_table
```

```python
ingredients_table.loc[0, 'ewg_ingre']
```

```python
import json

def preprocess_ewg(ewg):
    if ewg is None or ewg == "":
        return None 

    if isinstance(ewg, dict):
        ewg_dict = ewg
    elif isinstance(ewg, str):
        try:
            ewg_dict = json.loads(ewg.replace("'", '"'))  
        except json.JSONDecodeError:
            print(f"Invalid JSON format: {ewg}") 
            return None  
    else:
        return None 

    ewg_dict.pop("background", None)  
    ewg_dict.pop("url", None)  
    return ewg_dict

ingredients_table["preprocessed_ewg_ingre"] = ingredients_table["ewg_ingre"].apply(preprocess_ewg)
ingredients_table
```

```python
ingredients_table["preprocessed_ewg_ingre"] = ingredients_table["preprocessed_ewg_ingre"].apply(lambda x: str(x) if isinstance(x, dict) else x)
```
